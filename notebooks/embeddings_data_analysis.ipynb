```python
import numpy as np
import torch
import glob
import pandas as pd
# Load the PyTorch tensor from a file
x_train = torch.load('./raft_embeddings/0.pt')

paths = glob.glob("./calib_challenge_valdag_fork/labeled/*.txt")
data = pd.read_csv(paths[0], sep=" ", header=None)[:-1]
y_train = torch.tensor(data.values, dtype=torch.float32)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))  # 1 row, 2 columns

# Plot the NumPy array
ax1.plot(x_train.numpy())
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Our embedding')
ax1.legend()

# Plot the second subplot on ax2
ax2.plot(y_train.numpy())
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Ground Truth')
ax2.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the combined figure with both subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Find the minimum and maximum values in the tensor
min_value = x_train.min()
max_value = x_train.max()

# Normalize the tensor to the (0, 1) interval
normalized_tensor = (x_train - min_value) / (max_value - min_value)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))  # 1 row, 2 columns

# Plot the NumPy array
ax1.plot(normalized_tensor.numpy())
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Our embedding')
ax1.legend()

# Plot the second subplot on ax2
ax2.plot(y_train.numpy())
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Ground Truth')
ax2.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the combined figure with both subplots
plt.show()
```
